```python
from numpy import *
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
img=cv2.imread('/content/messi.jpg')
plt.imshow(img)
```

# Negative of Image

```python
#s=(L-1) - r
img_neg = (2**8) - img
plt.imshow(img_neg)
```

# Thresholding

```python
threshold = int(input("Enter threshold "))
s = np.where(img > threshold, 0, 255)
plt.imshow(s)
```

# Remove Background

```python
a = int(input("A "))
b = int(input("B" ))
a1 = img > a
b1 = img < b
q = np.where(a1 & b1, 0, 255)
plt.imshow(q)
```

# Remove subject

```python
a = int(input("A "))
b = int(input("B" ))
a2 = img > a
b2 = img < b
p = np.where(a2 & b2, img, 255)
plt.imshow(p)
```

```python
plt.subplot(2,2,1)
plt.title("Negative")
plt.imshow(img_neg)

plt.subplot(2,2,2)
plt.title("Thresholding")
plt.imshow(s)

plt.subplot(2,2,3)
plt.title("Without BG")
plt.imshow(q)

plt.subplot(2,2,4)
plt.title("With only BG")
plt.imshow(p)
```
